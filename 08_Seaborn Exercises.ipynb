## The Data
We will be working with a famous titanic data set for these exercises. 

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

# Exercises


```python
sns.jointplot(x="fare", y="age", data=titanic)
```

```python
sns.displot(titanic["fare"], color="red", bins=30, kde=False)
```

```python
sns.boxplot(x="class", y="age", data=titanic, palette="rainbow", legend=False, hue="class")
```

```python
sns.swarmplot(x="class", y="age", data=titanic, palette="coolwarm")
```

```python
sns.countplot(x="sex", data=titanic, palette="rainbow")
```

```python
sns.heatmap(titanic.corr(numeric_only=True), cmap="coolwarm")
```

```python
g = sns.FacetGrid(data=titanic, col="sex")
g.map(plt.hist,"age")
```
