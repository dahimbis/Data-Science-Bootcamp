## Lambda Expressions Basics

```python
def square(num):
    result = num**2
    return result
```

```python
square(100)
```

```python
def even(num):
    result = num%2
    return result==0
```

```python
even(5)
```

```python
even = lambda num: num%2==0
```

```python
even(500)
```

```python
first_char = lambda s:s[0]
```

```python
first_char("Ahimbisibwe")
```

```python
reverse = lambda s:s[::-1]
```

```python
reverse("Ahimbisibwe")
```
