## MATPLOTLIB

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(1,5,11)
```

```python
x
```

```python
y = x**2
```

```python
plt.plot(x,y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Plot of X VS Y")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,"r")
plt.subplot(1,2,2)
plt.plot(x,y,"b")
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.plot(y,x)
axes.set_xlabel("X")
axes.set_ylabel("Y")
axes.set_title("X VS Y")
```

```python
fig = plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.4,0.3,0.4])
axes1.plot(x,y,"r")
axes2.plot(y,x,"b")
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
plt.tight_layout()
for n in axes:
    n.plot(x,y)
axes[0].plot(x,y)
axes[1].plot(y,x)
axes[1].set_title("One")
```

```python
fig = plt.figure(figsize=(3,2))
ax= fig.add_axes([0,0,1,1,])
ax.plot(x,y)
```

```python
fig, axes=plt.subplots(nrows=2, ncols=1, figsize=(8,6))
axes[0].plot(x,y,"r")
axes[1].plot(y,x,"b")
plt.tight_layout()
fig.savefig("Image.png", dpi=200)
```
