```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop("benign_0__mal_1", axis=1).values
y=df["benign_0__mal_1"].values
```

```python
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train.shape
```

```python
model = Sequential()
model.add(Dense(30, activation="relu"))
model.add(Dense(15, activation="relu"))

model.add(Dense(1, activation ="sigmoid"))

model.compile(optimizer="adam", loss="binary_crossentropy")
```

```python
model.fit(x=X_train, y=y_train, validation_data=(X_test, y_test), epochs=500, batch_size=40)
```

```python
model_loss = pd.DataFrame(model.history.history)
```

```python
model_loss.plot()
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor="val_loss",mode="min", verbose=1, patience=25)
```

```python
model = Sequential()
model.add(Dense(30, activation="relu"))
model.add(Dense(15, activation="relu"))

model.add(Dense(1, activation ="sigmoid"))

model.compile(optimizer="adam", loss="binary_crossentropy")
```

```python
model.fit(x=X_train, y=y_train, validation_data=(X_test, y_test), epochs=600, callbacks=early_stop)
```

```python
model_loss = pd.DataFrame(model.history.history).plot()
```

```python
from tensorflow.keras.layers import Dropout
```

```python
model = Sequential()
model.add(Dense(30, activation="relu"))
model.add(Dropout(0.5))
model.add(Dense(15, activation="relu"))
model.add(Dropout(0.5))
model.add(Dense(1, activation ="sigmoid"))

model.compile(optimizer="adam", loss="binary_crossentropy")
```

```python
model.fit(x=X_train, y=y_train, validation_data=(X_test, y_test), epochs=500, batch_size=40)
```

```python
model_loss = pd.DataFrame(model.history.history).plot()
```
